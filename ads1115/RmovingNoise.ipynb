```python
import pandas as pd
import random
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy import stats
import glob
import os
from sklearn.preprocessing import LabelEncoder
import pickle
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
cwd = os.getcwd()
rel_path = "Sensing/Noise"

path = os.path.join(cwd, rel_path)
print("1   ",path)
all_files = glob.glob(path + "/*.csv")

print("2  ",all_files)
```

```python
noise='/Users/krishnamodi/Desktop/Lab/ads1115/Sensing/NOISE.csv'
dfnoise = pd.read_csv(noise, index_col=None, header=0)
dfnoise['avgX']=dfnoise['X'].rolling(window=20).mean()
dfnoise['avgY']=dfnoise['Y'].rolling(window=20).mean()
# print(df)
print("max=",dfnoise['Y'].max())
print("min=",dfnoise['Y'].min())
print('mean=',dfnoise['Y'].mean())
```

```python
li = []
i=0
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0 )
#     print(df)
    df = df[df['Y'] <= dfnoise['Y'].min()] 
    i=i+1
    print(i)
    print(df.mean(axis = 0,skipna = True))
    df['meanY']=df.mean(axis = 1,skipna = True)
    df['avgX']=df['X'].rolling(window=10).mean()
    df['avgY']=df['Y'].rolling(window=10).mean()
    df['minX']=df['X'].rolling(20).min()
    df['minY']=df['Y'].rolling(20).min()
    df.dropna()
    
    df.to_csv('Sensing/Noise/filtered/F'+str(i), sep='\t')
    print(df)
    li.append(df)
    #plotting the bar chart between the quality and its count
    
#     sb.countplot(x='Label',data=df)
    
#     plt.show()


frame = pd.concat(li, axis=0, ignore_index=True)
frame = frame.dropna()
frame
```
